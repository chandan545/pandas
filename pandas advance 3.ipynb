```python
#merge
df1 = pd.DataFrame({'key1':[1,2,4,5,6],
                    'key2':[4,5,6,7,8],
                    'key3':[3,4,5,6,6]
                   })
df1
```

```python
df2 = pd.DataFrame({'key1':[1,2,45,6,67],
                    'key4':[56,5,6,7,8],
                    'key5':[3,56,5,6,6]
                   })
df2
```

```python
pd.merge(df1, df2, how = 'inner') #inner merge
```

```python
pd.merge(df1, df2, how = 'left') #left merge
```

```python
pd.merge(df1, df2, how = 'right') #right merge
```

```python
pd.merge(df1, df2, how = 'outer') #all the common element of the column from both the data frames
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, how = "cross")
```

```python
pd.merge(df1, df2, how = 'left', left_on="key2", right_on="key4")
```
